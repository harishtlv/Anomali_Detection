```python
import matplotlib.pyplot as plt
import torch
import cv2

def display_anomaly_map_with_score_and_input(result_dict, image_path):
    # Extracting the anomaly map and pred_score from the dictionary
    anomaly_map = result_dict['anomaly_map'].squeeze().numpy()  # Convert tensor to numpy array
    pred_score = result_dict['pred_score'].item()  # Extracting scalar value from tensor

    # Reading the original input image
    original_image = cv2.imread(image_path)
    original_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)  # Convert BGR (OpenCV default) to RGB

    # Create a figure with two subplots: one for the original image and one for the anomaly map
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))

    # Plot the original input image
    axes[0].imshow(original_image)
    axes[0].set_title('Original Image')
    axes[0].axis('off')  # Hide axes

    # Plot the anomaly map
    im = axes[1].imshow(anomaly_map, cmap='summer', interpolation='nearest')
    axes[1].set_title(f'Anomaly Map\nPredicted Score: {pred_score:.2f}')
    axes[1].axis('off')  # Hide axes

    # Add a color bar for the anomaly map
    fig.colorbar(im, ax=axes[1], fraction=0.046, pad=0.04, label='Anomaly Score')

    # Display the plot
    plt.show()

image_path = "/content/drive/MyDrive/M_tech_proj/MVTech/carpet/test/color/000.png"
display_anomaly_map_with_score_and_input(predictions, image_path)
```

```python
import torch
from anomalib.models.image import Patchcore
from PIL import Image
from torchvision import transforms
import time
import matplotlib.pyplot as plt

# Initialize the model
model = Patchcore.load_from_checkpoint(
    checkpoint_path="/content/drive/MyDrive/Final_Project/Phase3/MVTec/carpet/v0/weights/lightning/model.ckpt"
)

# Convert model to TorchScript for optimization
# model = model.to_torchscript()

# Move the model to GPU and set to evaluation mode
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device).eval()

# Use FP16 precision if possible
if torch.cuda.is_available():
    model = model.half()

# Example image path
image_path = "/content/drive/MyDrive/Final_Project/MVTech/carpet/test/color/000.png"
image = Image.open(image_path)

# Convert PIL Image to Tensor and apply the necessary transformations using functional API
image = F.resize(image, (256, 256))  # Resize image to 256x256
image = F.to_tensor(image)  # Convert to tensor
image = F.normalize(image, mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  # Normalize the tensor

# Add batch dimension
image = image.unsqueeze(0)

# Use FP16 precision for the input image if model is in FP16
if torch.cuda.is_available():
    image = image.half()

# Move the image to GPU
image = image.to(device)

# Perform a warm-up run (optional, but can reduce first-time inference latency)
with torch.no_grad():
    _ = model(image)

# Perform the prediction with timing
start_time = time.time()
with torch.no_grad():
    predictions = model(image)
end_time = time.time()

def display_anomaly_map_with_score(result_dict):
    anomaly_map = result_dict['anomaly_map'].squeeze().cpu().numpy()  # Convert tensor to numpy array
    pred_score = result_dict['pred_score'].item()  # Extracting scalar value from tensor

    plt.imshow(anomaly_map, interpolation='nearest')
    plt.colorbar(label='Anomaly Score')
    plt.title(f'Predicted Score: {pred_score:.2f}')
    plt.show()

display_anomaly_map_with_score(predictions)
print(f"Time taken: {end_time - start_time:.4f} seconds")
```
